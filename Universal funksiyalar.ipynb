```python
import numpy as np
```

```python
# Unary funksiyalar (bitta argument qabul qiluvchi funksiya)
```

```python
arr = np.arange(10)
```

```python
arr
```

```python
np.sqrt(arr) #arr nomli massivning barcha elementlaridan kvadrat ildiz qaytaradi
```

```python
np.square(arr) # har bir elementning kvadratini qaytaradi
```

```python
np.exp(arr) # barcha elementlarning eksponentini qaytaradi
```

```python
np.log(arr[1:]) #  0-elementidan boshqa barcha elementlarning logarifmik qiymatini qaytaradi
```

```python
arr1 = np.random.randn(6) # 6 ta elementdan iborat taxminiy qiymatlarga ega massiv
```

```python
arr1
```

```python
qoldiq, butun = np.modf(arr1) # har bir elementni butun va qoldiq qismga ajratadi
```

```python
print(qoldiq)
print(butun)
```

```python
np.sign(arr1) # qiymatlarni -1 va 1 qiymatlarga ajratadi
```
