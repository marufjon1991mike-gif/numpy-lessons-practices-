```python
import numpy as np
```

```python
arr = np.arange(6)
arr
```

```python
np.save("data", arr) # massivni "data" nomi bilan saqlash
```

```python
arr2 = np.load("data.npy") # saqlab qo'ygan massivni choqirish
```

```python
arr3 = np.arange(20)
arr3
```

```python
np.savez("arrays", a = arr, b = arr3) #ikta massivni saqlash
```

```python
arrays = np.load("arrays.npz")
```

```python
arrays['a'] # a nom bilan saqlangan arr massivi
```
