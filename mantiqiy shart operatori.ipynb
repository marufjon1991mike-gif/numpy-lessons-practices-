```python
xarr = np.array([1.1, 1.2, 1.3, 1.4, 1.5])
```

```python
yarr = np.array([2.1, 2.2, 2.3, 2.4, 2.5])
```

```python
xarr
```

```python
yarr
```

```python
cond = np.array([True, False, True, True, False]) # agar element True bo'lsa xarr dan qiymat qabul qiladi, agar False bo'lsa yarr dan qiymat qabul qiladi
```

```python
# where
result = np.where(cond, xarr, yarr) #agar element True bo'lsa xarr dan qiymat qabul qiladi, agar False bo'lsa yarr dan qiymat qabul qiladi
```

```python
result
```
